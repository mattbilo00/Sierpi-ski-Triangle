```python
# Imports
import random
import matplotlib.pyplot as plt

# Ask user for number of points to plot
user_num = input("Enter the number of points to plot: ")

# Value specifications for the equilateral trianlge
length = 1000   # requires: length>= 4
half_length = length / 2
top_point = ((3/4) * (length)**2)**(1/2)
tri = [[0,0], [length, 0], [half_length, top_point]]

# Equations for the sides of the triangle
def y_up(x):
    y = (3**(1/2)) * x
    return y

def y_down(x):
    y = -y_up(x) + (2 * top_point)
    return y

# Make a radnom point in the triangle
y = 0
x = random.randrange(0, length)
if x < half_length:
    y = random.randrange(0, int(y_up(x)))
else:
    y = random.randrange(0, int(y_down(x)))

points_x = [x]
points_y = [y]

# Make the rest of the points
for i in range(0, int(user_num)):
    # Choose a random triangle corner
    tri_point = random.randrange(0, 3)
    
    # Find the average point between the most recent point and the random corner
    x = (points_x[i] + tri[tri_point][0]) / 2
    y = (points_y[i] + tri[tri_point][1]) / 2
    points_x.append(x)
    points_y.append(y)
    
# Plot the points
plt.figure(figsize = (10, 10))
plt.plot(points_x, points_y, 'o', markersize = 0.5)
plt.xlim(-0.01*top_point, 1.01*length)
plt.ylim(-0.01*top_point, 1.01*top_point)
plt.title("Sierpiski Triangle")
plt.tight_layout()
```
